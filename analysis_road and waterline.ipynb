```python
import xmltodict
import pandas as pd

# File paths
xml_files = {
    "Main Road": "Datasets_Hackathon 1/Datasets_Hackathon/Streamwater_Line_Road_Network/Main_Road.shp.xml",
    "Streamwater": "Datasets_Hackathon 1/Datasets_Hackathon/Streamwater_Line_Road_Network/Streamwater.shp.xml"
}

def parse_xml(file_path):
    """Parse an XML file and convert it to a dictionary"""
    with open(file_path, "r", encoding="utf-8") as file:
        xml_content = file.read()
    return xmltodict.parse(xml_content)

# Extract information from each file
metadata_info = []

for name, file_path in xml_files.items():
    data = parse_xml(file_path)

    # Extract key metadata fields
    metadata = {
        "Dataset Name": name,
        "Creation Date": data["metadata"]["Esri"]["CreaDate"],
        "Coordinate System": data["metadata"]["Esri"]["DataProperties"]["coordRef"]["geogcsn"],
        "Projection": data["metadata"]["Esri"]["DataProperties"]["coordRef"]["peXml"],
        "Process History": data["metadata"]["Esri"]["DataProperties"]["lineage"]["Process"],
        "Attributes": [attr["attrlabl"] for attr in data["metadata"]["eainfo"]["detailed"]["attr"]]
    }
    
    metadata_info.append(metadata)

# Convert to DataFrame for better visualization
df_metadata = pd.DataFrame(metadata_info)

# Display the results
df_metadata
```

```python
# Define Assaba region (adjust coordinates if necessary)
assaba_region = ee.Geometry.Rectangle([18.398501, -13.042211, 15.036111, -10.369845])  # Modify as needed

# Load MODIS Land Cover dataset (2000 - 2023)
modis_landcover = ee.ImageCollection("MODIS/006/MCD12Q1").select('LC_Type1')

# Filter dataset from 2000 to 2023 for Assaba
modis_filtered = modis_landcover.filterDate('2000-01-01', '2023-12-31').map(lambda img: img.clip(assaba_region))
```

```python
import ee
import pandas as pd
import matplotlib.pyplot as plt

# Define land cover class codes
landcover_classes = {
    1: "Evergreen Needleleaf Forest",
    2: "Evergreen Broadleaf Forest",
    3: "Deciduous Needleleaf Forest",
    4: "Deciduous Broadleaf Forest",
    5: "Mixed Forests",
    7: "Open Shrublands",
    8: "Woody Savannas",
    9: "Savannas",
    10: "Grasslands",
    12: "Croplands",
    16: "Barren Land"
}

# Extract land cover trends per class
years = list(range(2001, 2021))
cover_trends = {str(k): [] for k in landcover_classes.keys()}

for year in years:
    image = modis_filtered.filterDate(f'{year}-01-01', f'{year}-12-31').first()

    if image:
        for class_code in landcover_classes.keys():
            # Create a mask for each class
            class_mask = image.eq(class_code)
            class_area = class_mask.multiply(ee.Image.pixelArea()).reduceRegion(
                reducer=ee.Reducer.sum(),
                geometry=assaba_region,
                scale=500,
                maxPixels=1e9
            ).getInfo()

            # Store values
            cover_trends[str(class_code)].append(class_area.get("LC_Type1", 0) / 1e6)  # Convert m to km
    else:
        for class_code in landcover_classes.keys():
            cover_trends[str(class_code)].append(0)

# Convert to DataFrame
df_landcover = pd.DataFrame(cover_trends, index=years)

# Plot the trends
plt.figure(figsize=(12, 6))

for class_code, label in landcover_classes.items():
    plt.plot(df_landcover.index, df_landcover[str(class_code)], label=label)

plt.xlabel("Year")
plt.ylabel("Area (sq km)")
plt.title("Land Cover Change in Assaba (2001-2020)")
plt.legend()
plt.grid()
plt.show()
```

```python
import math
import os
import rasterio
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
from rasterio.plot import show
from rasterio.mask import mask
from collections import defaultdict
import imageio

#  Set the folder containing MODIS Land Cover TIFF files
data_folder = "Datasets_Hackathon 1/Datasets_Hackathon/Modis_Land_Cover_Data"  # Change this to your folder path

#  Define Land Cover Classes from MODIS MCD12Q1
land_cover_classes = {
    0: "Water",
    1: "Evergreen Needleleaf Forest",
    2: "Evergreen Broadleaf Forest",
    3: "Deciduous Needleleaf Forest",
    4: "Deciduous Broadleaf Forest",
    5: "Mixed Forests",
    6: "Closed Shrublands",
    7: "Open Shrublands",
    8: "Woody Savannas",
    9: "Savannas",
    10: "Grasslands",
    11: "Permanent Wetlands",
    12: "Croplands",
    13: "Urban and Built-up",
    14: "Cropland/Natural Vegetation Mosaic",
    15: "Barren",
    16: "Unclassified"
}

#  Store Data
land_cover_data = {}  # Stores raster arrays per year
land_cover_stats = {}  # Stores land cover distribution per year
years = []  # Store available years

#  Load all TIFF files in the folder
for filename in sorted(os.listdir(data_folder)):
    if filename.endswith(".tif"):  # Process only .tif files
        year = int(filename[:4])  # Extract the year from filename (e.g., "2010LCT.tif")
        years.append(year)
        file_path = os.path.join(data_folder, filename)

        with rasterio.open(file_path) as src:
            land_cover_data[year] = src.read(1)  # Read the first band (land cover classes)
            profile = src.profile  # Get metadata

            # Compute statistics for land cover distribution
            unique, counts = np.unique(land_cover_data[year], return_counts=True)
            land_cover_stats[year] = dict(zip(unique, counts))

#  Display Dataset Metadata
print("Dataset Metadata:")
print(f"CRS: {profile['crs']}")
print(f"Resolution: {profile['transform'][0]} meters per pixel")
print(f"Bounding Box: {profile['transform']}")
print(f"Years Available: {years}")

num_years = len(years)
cols = 4  # Number of columns in the grid (adjust as needed)
rows = math.ceil(num_years / cols)  # Automatically calculate required rows

#  Create a grid of subplots
fig, axes = plt.subplots(rows, cols, figsize=(cols * 4, rows * 4))  # Adjust figure size

#  Flatten axes array for easy iteration
axes = axes.flatten()

#  Loop through each year and plot land cover data
for i, year in enumerate(years):
    ax = axes[i]
    ax.set_title(f"{year}", fontsize=10)
    im = ax.imshow(land_cover_data[year], cmap="tab20", interpolation="nearest")
    ax.axis("off")  # Hide axis for cleaner visualization

#  Remove any extra empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

#  Add a colorbar (common for all subplots)
cbar_ax = fig.add_axes([0.92, 0.2, 0.02, 0.6])  # Position the colorbar
fig.colorbar(im, cax=cbar_ax, label="Land Cover Class")

#  Show final grid layout
plt.suptitle("MODIS Land Cover Maps Over Time", fontsize=14)
plt.tight_layout(rect=[0, 0, 0.9, 1])  # Adjust layout to fit colorbar
plt.show()

#  Analyze Land Cover Change Over Time
land_cover_trends = defaultdict(list)

for year in years:
    for lc_class, count in land_cover_stats[year].items():
        land_cover_trends[lc_class].append((year, count))

#  Plot Trends for Key Land Cover Classes
plt.figure(figsize=(12, 6))
for lc_class, trend in land_cover_trends.items():
    years_list, values = zip(*trend)
    plt.plot(years_list, values, label=land_cover_classes.get(lc_class, f"Class {lc_class}"))

plt.xlabel("Year")
plt.ylabel("Land Cover Area (Pixel Count)")
plt.title("Land Cover Change Trends (MODIS)")
plt.legend(loc="upper right", fontsize="small")
plt.grid()
plt.show()

#  Create Land Cover Animation Over Time
gif_frames = []
for year in years:
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.set_title(f"Land Cover Change ({year})")
    im = ax.imshow(land_cover_data[year], cmap="tab20", interpolation="nearest")
    plt.colorbar(im, ax=ax)
    
    # Save each frame
    temp_path = f"frame_{year}.png"
    plt.savefig(temp_path)
    plt.close()
    gif_frames.append(imageio.imread(temp_path))
    os.remove(temp_path)  # Remove temp files after saving

#  Save GIF Animation
imageio.mimsave("land_cover_change.gif", gif_frames, duration=1)
print(" Land Cover Change GIF saved as 'land_cover_change.gif'")

#  Generate Summary Statistics
summary_stats = {}

for year, stats in land_cover_stats.items():
    total_pixels = sum(stats.values())
    summary_stats[year] = {land_cover_classes.get(k, f"Class {k}"): (v / total_pixels) * 100 for k, v in stats.items()}

import pandas as pd
df_summary = pd.DataFrame(summary_stats).T
df_summary
```

```python
import math
import os
import rasterio
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
from rasterio.plot import show
from collections import defaultdict
import imageio

#  Set the folder containing MODIS Land Cover TIFF files
data_folder = "Datasets_Hackathon 1/Datasets_Hackathon/Modis_Land_Cover_Data"  # Change this to your folder path

#  Define Correct MODIS Land Cover Classes
land_cover_classes = {
    0: "Water",
    1: "Evergreen Needleleaf Forest",
    2: "Evergreen Broadleaf Forest",
    3: "Deciduous Needleleaf Forest",
    4: "Deciduous Broadleaf Forest",
    5: "Mixed Forests",
    6: "Closed Shrublands",
    7: "Open Shrublands",
    8: "Woody Savannas",
    9: "Savannas",
    10: "Grasslands",
    11: "Permanent Wetlands",
    12: "Croplands",
    13: "Urban and Built-up",
    14: "Cropland/Natural Vegetation Mosaic",
    15: "Barren",
    16: "Unclassified"  # Use for unknowns
}

#  Remove Invalid Classes (-128, 255)
valid_classes = set(land_cover_classes.keys())  # Only allow known class values

#  Store Data
land_cover_data = {}  # Stores raster arrays per year
land_cover_stats = {}  # Stores land cover distribution per year
years = []  # Store available years

#  Load all TIFF files in the folder
for filename in sorted(os.listdir(data_folder)):
    if filename.endswith(".tif"):  # Process only .tif files
        year = int(filename[:4])  # Extract the year from filename (e.g., "2010LCT.tif")
        years.append(year)
        file_path = os.path.join(data_folder, filename)

        with rasterio.open(file_path) as src:
            raw_data = src.read(1)  # Read the first band (land cover classes)
            
            #  Remove invalid values (-128, 255, etc.)
            cleaned_data = np.where(np.isin(raw_data, list(valid_classes)), raw_data, np.nan)
            land_cover_data[year] = cleaned_data

            #  Compute statistics for land cover distribution
            unique, counts = np.unique(cleaned_data[~np.isnan(cleaned_data)], return_counts=True)
            land_cover_stats[year] = dict(zip(unique, counts))

#  Display Dataset Metadata
print("Dataset Metadata:")
print(f"Years Available: {years}")

#  Grid Layout for Visualization
num_years = len(years)
cols = 4  # Number of columns in the grid (adjust as needed)
rows = math.ceil(num_years / cols)  # Automatically calculate required rows

#  Create a grid of subplots
fig, axes = plt.subplots(rows, cols, figsize=(cols * 4, rows * 4))  # Adjust figure size
axes = axes.flatten()

#  Loop through each year and plot land cover data
for i, year in enumerate(years):
    ax = axes[i]
    ax.set_title(f"{year}", fontsize=10)
    im = ax.imshow(land_cover_data[year], cmap="tab20", interpolation="nearest")
    ax.axis("off")  # Hide axis for cleaner visualization

#  Remove any extra empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

#  Add a colorbar (common for all subplots)
cbar_ax = fig.add_axes([0.92, 0.2, 0.02, 0.6])  # Position the colorbar
fig.colorbar(im, cax=cbar_ax, label="Land Cover Class")

#  Add Class Labels to Legend
legend_patches = [plt.Line2D([0], [0], color=plt.cm.tab20(i / len(valid_classes)), lw=4, label=label)
                  for i, (lc, label) in enumerate(land_cover_classes.items()) if lc in valid_classes]
plt.legend(handles=legend_patches, loc='lower right', fontsize="small", title="Land Cover Classes")

#  Show final grid layout
plt.suptitle("MODIS Land Cover Maps Over Time (Without Invalid Classes)", fontsize=14)
plt.tight_layout(rect=[0, 0, 0.9, 1])  # Adjust layout to fit colorbar

#  Analyze Land Cover Change Over Time
land_cover_trends = defaultdict(list)

for year in years:
    for lc_class, count in land_cover_stats[year].items():
        land_cover_trends[lc_class].append((year, count))

#  Plot Trends for Key Land Cover Classes
plt.figure(figsize=(12, 6))
for lc_class, trend in land_cover_trends.items():
    years_list, values = zip(*trend)
    plt.plot(years_list, values, label=land_cover_classes.get(lc_class, f"Class {lc_class}"))

plt.xlabel("Year")
plt.ylabel("Land Cover Area (Pixel Count)")
plt.title("Land Cover Change Trends (MODIS)")
plt.legend(loc="upper right", fontsize="small")
plt.grid()
plt.show()

#  Create Land Cover Animation Over Time
gif_frames = []
for year in years:
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.set_title(f"Land Cover Change ({year})")
    im = ax.imshow(land_cover_data[year], cmap="tab20", interpolation="nearest")
    plt.colorbar(im, ax=ax)
    
    # Save each frame
    temp_path = f"frame_{year}.png"
    plt.savefig(temp_path)
    plt.close()
    gif_frames.append(imageio.imread(temp_path))
    os.remove(temp_path)  # Remove temp files after saving

#  Save GIF Animation
imageio.mimsave("land_cover_change.gif", gif_frames, duration=1)
print(" Land Cover Change GIF saved as 'land_cover_change.gif'")

#  Generate Summary Statistics
summary_stats = {}

for year, stats in land_cover_stats.items():
    total_pixels = sum(stats.values())
    summary_stats[year] = {land_cover_classes.get(k, f"Class {k}"): (v / total_pixels) * 100 for k, v in stats.items()}

import pandas as pd
df_summary = pd.DataFrame(summary_stats).T
df_summary
```
